```python
from ultralytics import YOLO
import cv2 as cv

model = YOLO("YOLO_model.pt")

cap = cv.VideoCapture(0) 

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    person_count = 0
    for result in results:
        for cls in result.boxes.cls:
            if int(cls) == 0:
                person_count += 1

    annotated_frame = results[0].plot()

    cv.putText(annotated_frame, f'People Count: {person_count}', (10, 30),
                cv.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)
    cv.imshow("Crowd Counter", annotated_frame)

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv.destroyAllWindows()
```
